```python
wt = wt[wt['cell volume (um^3)'] > 20]
wt = wt[wt['artifacts'] < 5]

print('wt: ' + str(len(wt.cellID.unique())))

hxk2 = hxk2[hxk2['cell volume (um^3)'] > 20]
hxk2 = hxk2[hxk2['artifacts'] < 5]

print('hxk2∆: ' + str(len(hxk2.cellID.unique())))

snf1 = snf1[snf1['cell volume (um^3)'] > 20]
snf1 = snf1[snf1['artifacts'] < 5]

print('snf1∆: ' + str(len(snf1.cellID.unique())))

hxk2snf1 = hxk2snf1[hxk2snf1['cell volume (um^3)'] > 20]
hxk2snf1 = hxk2snf1[hxk2snf1['artifacts'] < 5]

print('hxk2∆ snf1∆: ' + str(len(hxk2snf1.cellID.unique())))


reg1 = reg1[reg1['cell volume (um^3)'] > 20]
reg1 = reg1[reg1['artifacts'] < 5]

print('reg1∆: ' + str(len(reg1.cellID.unique())))

rgt2snf3 = rgt2snf3[rgt2snf3['cell volume (um^3)'] > 20]
rgt2snf3 = rgt2snf3[rgt2snf3['artifacts'] < 5]

print('rgt2∆ snf3∆: ' + str(len(rgt2snf3.cellID.unique())))

mig1mig2 = mig1mig2[mig1mig2['cell volume (um^3)'] > 20]
mig1mig2 = mig1mig2[mig1mig2['artifacts'] < 5]

print('mig1∆ mig2∆: ' + str(len(mig1mig2.cellID.unique())))


gal = gal[gal['cell volume (um^3)'] > 20]
gal = gal[gal['artifacts'] < 5]

print('glucose > galactose: ' + str(len(gal.cellID.unique())))

prot_N = prot_N[prot_N['cell volume (um^3)'] > 20]
prot_N = prot_N[prot_N['artifacts'] < 5]

print('prototroph -N: ' + str(len(prot_N.cellID.unique())))

prot_C = prot_C[prot_C['cell volume (um^3)'] > 20]
prot_C = prot_C[prot_C['artifacts'] < 5]

print('prototroph -C: ' + str(len(prot_C.cellID.unique())))
```

## Recoverers v arresters: dynamics

```python
times_full = np.arange(-60, 600 + 15, 15)
```

```python
times_full
```

```python
long_IDs = wt[wt['time'] == '600'].cellID.unique()
```

```python
wt_long = wt[wt['cellID'].isin(long_IDs)]
wt_long = wt_long[wt_long['cellID'].isin(wt[wt['time'] == '0'].cellID.unique())]
```

```python
len(wt_long.cellID.unique())
```

```python
wt_collapsed = []
wt_recovered = []
for t in times_full:
    temp_df = wt_long[wt_long['time'] == str(t)]
    wt_collapsed.append(len(temp_df[temp_df['volume-weighted average sphericity of mitochondrion'] > 0.7]))
    wt_recovered.append(len(temp_df[temp_df['volume-weighted average sphericity of mitochondrion'] <= 0.7]))
```

```python
sns.set_context('paper', font_scale = 1.5)
sns.set_style('ticks')

f, ax = plt.subplots()

plt.axvline(x = 0, color = 'black', linestyle = (0, (3,5)))

sns.set_context(rc = {'lines.linewidth': 2.0})

plt.plot(times_full[::2], wt_collapsed[::2], marker = '.', markersize = 10, color = grays[4],
        label = 'Arresters')
plt.plot(times_full[::2], wt_recovered[::2], marker = '.', markersize = 10, color = reds[5],
        label = 'Recoverers')

ax.set_ylim(0, 300)
ax.set_ylabel('Number of cells')
ax.set_xlabel('Time (min)')

ax.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0, frameon = False);

sns.despine()
```
